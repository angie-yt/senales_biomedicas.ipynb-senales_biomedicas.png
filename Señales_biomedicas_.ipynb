```python
import matplotlib.pyplot as plt

#Datos de singnos vitales simulados

tiempo = [1, 2, 3, 4, 5 ] #minutos
frecuencia = [95, 98, 102, 99, 105] #latidos por minuto
presion_sistolica = [120, 122, 130, 128, 135] #mmhg


# Graficamos
plt.figure(figsize=(8,4))
plt.plot(tiempo, frecuencia, label="Frecuencia Cardíaca (bpm)")
plt.plot(tiempo, presion_sistolica, label="Presión Sistólica (mmHg)")

# Estética del gráfico
plt.xlabel("Tiempo (min)")
plt.ylabel("Valor")
plt.title("Señales Biomédicas")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Gráfica mejorada
plt.figure(figsize=(8, 4))
plt.plot(tiempo, frecuencia, marker="o", linestyle="-", linewidth=2, label="Frecuencia Cardíaca (bpm)")
plt.plot(tiempo, presion_sistolica, marker="s", linestyle="--", linewidth=2, label="Presión Sistólica (mmHg)")
plt.xlabel("Tiempo (min)")
plt.ylabel("Valor")
plt.title("Señales Biomédicas Mejoradas")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("senales_biomedicas.png")
plt.show()
```
